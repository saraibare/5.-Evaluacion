## **EUA**

```python
euay = eua.select_dtypes(include='number')
euay.head()
```

```python
euax = eua.select_dtypes(include='object')
euax.head()
```

### **Valores nulos**

Primeramente, se separó el dataset para tener dos conjuntos que mostraran las variables categóricas y numéricas. A los valores nulos de las variables categóricas, se reemplazó con otra categoría "Sin datos", con el fin de no alterar los resultados. Por otra parte, en cuanto a las variables numéricas, se eliminaron las columnas que carecían de algún dato y las sobrantes con nulos, se reemplazó con la media. Esto es debido a que, si elegimos reemplazar con 0, estaríamos suponiendo que no tiene esa característica, por lo que podría alterar el resultado. De otra forma, si se llegase a utilizar los métodos backfill o forwardfill, al ser los nulos más de una sexta parte del total, estaríamos incurriendo a un resultado bastante sesgado. Por ende, se eligió una medida de tendencia central para no afectar tanto la distribución.  

```python
#Verificar la cantidad de valores nulos por columna
euax.isnull().sum()
```

```python
#Sustitución de valores nulos de todo el dataframe por una frase
euax=euax.fillna('Sin datos')
euax.isnull().sum()
```

```python
euay.isnull().sum()
```

```python
#Se eliminan las columnas sin datos
euay=euay.drop(['neighbourhood_group_cleansed','bathrooms','calendar_updated'], axis=1)
#Sustitución de valores nulos con promedio o media
euay=euay.fillna(round(euay.mean(),1))
```

```python
euay.isnull().sum()
```

### **Outliers**

Se utilizó el Método de rango interquartílico (IQR) debido a que suele comprender un rango más acortado que aquel usando el de desviación estándar, permitiendo así eliminar varios valores atípicos. Además, es el método más utilizado, puesto que se emplea en la creación de gráficas. 

```python
percentile25=euay.quantile(0.25) #Q1
percentile75=euay.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr

#Obtenemos datos limpios del Dataframe
euay_iqr= euay[(euay<=Limite_Superior_iqr)&(euay>=Limite_Inferior_iqr)]
```

```python
euay_iqr.isnull().sum()
```

```python
euay_iqr = euay_iqr.fillna(round(euay_iqr.mean(),1))
euay_iqr.isnull().sum()
```

```python
eua = pd.concat([euax,euay_iqr], axis=1)
eua.head()
```

# **ETAPA 3: Extracción de Características y Visualización de DATOS**

## **host_response_time**

```python
#!pip install funpymodeling
from funpymodeling.exploratory import freq_tbl 

tableAeua= freq_tbl(eua['host_response_time'])
tableAeua = tableAeua.drop(['cumulative_perc','percentage'], axis=1)
tableAeua = tableAeua.rename(columns = {'frequency':'EUA'})

tableAmexico= freq_tbl(mexico['host_response_time'])
tableAmexico = tableAmexico.drop(['cumulative_perc','percentage'], axis=1)
tableAmexico = tableAmexico.rename(columns = {'frequency':'MEX',})

tableAespaña= freq_tbl(españa['host_response_time'])
tableAespaña = tableAespaña.drop(['cumulative_perc','percentage'], axis=1)
tableAespaña = tableAespaña.rename(columns = {'frequency':'ESP'})
```

```python
tableA = pd.merge(tableAeua, tableAmexico, on= 'host_response_time')
tableA = pd.merge(tableA, tableAespaña, on= 'host_response_time')
tableA = tableA.set_index('host_response_time')
tableA = tableA.drop(['Sin datos'], axis = 0)
tableA
```

```python
tableA.plot(kind = 'bar', color =  ['#c63637','#ff6961','#ffbfb0'], width=0.5, figsize=(10,4))
plt.xticks(rotation=0)
plt.title('Tiempo de respuesta')
plt.xlabel('Tiempo')
plt.ylabel('Frecuencia')
```
